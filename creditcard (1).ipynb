```python
fraud_trans = credit[credit['Class'] == 1]
normal_trans = credit[credit['Class'] == 0]
```

```python
print(fraud_trans.shape)
print(normal_trans.shape)
```

```python
normal_trans = normal_trans.sample(104)
print(normal_trans.shape)
```

```python
data = pd.concat([fraud_trans, normal_trans], axis = 0)
```

```python
data.shape
```

```python
data.sample(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x = data.drop('Class', axis = 1)
y = data['Class']
```

```python
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, stratify = y, random_state = 2529)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
model = LogisticRegression(max_iter= 500)
```

```python
model.fit(x_train, y_train)
```

```python
y_train_pred = model.predict(x_train)
trained_accuracy=accuracy_score(y_train_pred, y_train)
print(trained_accuracy)
```

```python
y_test_pred = model.predict(x_test)
test_accuracy=accuracy_score(y_test_pred, y_test)
print(test_accuracy)
```
